```python
#importing necessary libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Importing the dataset "Carbon dioxide (CO2) Emissions without Land Use, Land-Use Change and Forestry (LULUCF), in kilotonne CO2 equivalent"
#From http://data.un.org/Data.aspx?d=GHG&f=seriesID%3aCO2
df = pd.read_excel('Carbon Dioxide Emissions.xls')
df.head()
```

```python
#Data plotted at one time without any filters.
plt.plot(df['Year'], df['Value'])
plt.title('CO2 Emissions by Country')
plt.xlabel('Year')
plt.ylabel('Emissions')
plt.show()
```

```python
#Showing Australia CO2 Emissions from 1990 to 2015. From Graph there is an increase
aust_df = df[df['Country or Area'] == 'Australia']
plt.plot(aust_df['Year'], aust_df['Value'], 'r+')
plt.title('Australia CO2 Emissions')
plt.xlabel('Year')
plt.ylabel('Emissions Value')
plt.show()
```

```python
#Showing USA CO2 Emissions from 1990 to 2015. From Graph there has been a decrease in values from 2007 and on.
usa_df = df[df['Country or Area'] == 'United States of America']
plt.plot(usa_df['Year'], usa_df['Value'], 'r+')
plt.title('USA CO2 Emissions')
plt.xlabel('Year')
plt.ylabel('Emissions Value')
plt.show()
```

```python
#Showing Australia vs USA CO2 Emissions. Seen in this graph, the data shows Australian emissions levels as less than USA emissions 
from #1990 to 2015.

plt.plot(aust_df['Year'], aust_df['Value'], 'r+')
plt.plot(usa_df['Year'], usa_df['Value'], 'bo')
plt.title('Australia vs USA CO2 Emissions')
plt.xlabel('Year')
plt.ylabel('Emissions Value')
plt.legend(['Australia', 'USA'])
plt.show()
```

```python
#Creating the same dataset but this time as a boxplot showing the range of values from the USA and Australia
plt.figure(figsize = (15, 5))
y1 = aust_df['Value']
y2 = usa_df['Value']

plt.subplot(1, 2, 1)
plt.boxplot(y1)
plt.ylabel('CO2 Value')
plt.title('Australia')

plt.subplot(1, 2, 2)
plt.boxplot(y2)
plt.ylabel('CO2 Value')
plt.title('USA')

plt.show()
```
